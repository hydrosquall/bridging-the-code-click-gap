```python
from IPython.display import display

# A helper to render graphs inline
# REF: Mimetype code: https://ipywidgets.readthedocs.io/en/latest/examples/Output%20Widget.html?highlight=display#Output-widgets:-leveraging-Jupyter's-display-system
def displayVegaLite(spec): 
    display({'application/vnd.vegalite.v3+json': spec }, raw=True)
```

```python
# Simple Vega example from the vega-lite documentation

vegaLiteSpec = {
  "$schema": "https://vega.github.io/schema/vega-lite/v5.json",
  "description": "A simple bar chart with embedded data.",
  "data": {
    "values": [
      {"a": "A", "b": 28}, {"a": "B", "b": 55}, {"a": "C", "b": 43},
      {"a": "D", "b": 91}, {"a": "E", "b": 81}, {"a": "F", "b": 53},
      {"a": "G", "b": 19}, {"a": "H", "b": 87}, {"a": "I", "b": 52}
    ]
  },
  "mark": "bar",
  "encoding": {
    "x": {"field": "a", "type": "nominal", "axis": {"labelAngle": 0}},
    "y": {"field": "b", "type": "quantitative"}
  }
}
```

```python
# https://jupyterlab.readthedocs.io/en/stable/user/file_formats.html?highlight=vega-lite#vega-vega-lite
displayVegaLite(vegaLiteSpec)
```

```python
# Works with remote data too!
# https://vega.github.io/vega-lite/examples/rect_heatmap_weather.html

heatmapSpec = {
  "$schema": "https://vega.github.io/schema/vega-lite/v5.json",
  "data": {
      "url": "https://vega.github.io/vega-lite/data/seattle-weather.csv"
  },
  "title": "Daily Max Temperatures (C) in Seattle, WA",
  "config": {
      "view": {
          "strokeWidth": 0,
          "step": 13
      },
      "axis": {
          "domain": False
      }
  },
  "mark": "rect",
  "encoding": {
      "x": {
          "field": "date",
          "timeUnit": "date",
          "type": "ordinal",
          "title": "Day",
          "axis": {
              "labelAngle": 0,
              "format": "%e"
          }
      },
      "y": {
          "field": "date",
          "timeUnit": "month",
          "type": "ordinal",
          "title": "Month"
      },
      "color": {
          "field": "temp_max",
          "aggregate": "max",
          "type": "quantitative",
          "legend": {
              "title": None
          },
          "scale": { "scheme": "oranges" }
      }
  }
}

displayVegaLite(heatmapSpec)
```
